## II.  Introduction

La Fast fashion ou « mode rapide » fait référence à «des vêtements fabriqués à bas prix et à des prix qui copient les derniers styles de défilés et sont rapidement pompés dans les magasins afin de maximiser les tendances actuelles».

Le modèle de la mode rapide est appelé ainsi parce qu'il implique la conception, la production, la distribution et la commercialisation rapides de vêtements, ce qui signifie que les détaillants sont en mesure de tirer de grandes quantités d'une plus grande variété de produits et permettent aux consommateurs d'obtenir plus de mode et de différenciation des produits à un moment donné avec des prix bas.

L'impact environnemental de la mode rapide comprend l'épuisement des sources non renouvelables, l'émission de gaz à effet de serre, l'utilisation massive d'eau et d'énergie et plein d'autres impacts. L'industrie de la mode est la deuxième plus grande industrie consommatrice d'eau, nécessitant environ 700 gallons pour produire une chemise en coton et 2 000 gallons d'eau pour produire une paire de jeans. [_Business Insider_](https://www.businessinsider.com/fast-fashion-environmental-impact-pollution-emissions-waste-water-2019-10?r=US&IR=T#fashion-causes-water-pollution-problems-too-textile-dyeing-is-the-worlds-second-largest-polluter-of-water-since-the-water-leftover-from-the-dyeing-process-is-often-dumped-into-ditches-streams-or-rivers-19) prévient également que la teinture des textiles est le deuxième plus grand pollueur d'eau au monde, car l'eau restante du processus de teinture est souvent déversée dans des fossés, des ruisseaux ou des rivières.

Des solutions sont donc à rechercher pour une mode plus éthique et durable.


Il se pose donc la question de connaitre les habitudes de consommation fast fashion,le degré de connaissance des impacts de la fast fashion sur l’environnement et enfin de savoir comment inciter ces consommateurs à adopter une mode plus éthique et durable.

Cela indique que pour encourager les consommateurs à adopter des choix plus durables, il est essentiel de proposer des options de mode abordables et respectueuses de l'environnement, qui répondent également aux tendances actuelles de la mode. Les consommateurs sont à la recherche de vêtements qui leur permettent de se sentir à la mode et tendance, sans pour autant sacrifier leur portefeuille.

Cela souligne l'importance d'éduquer les consommateurs sur les choix de mode durables et de sensibiliser sur les impacts environnementaux de la fast fashion. Les marques comme ZARA, H&M, Bershka, SHEIN... peuvent également jouer un rôle important en proposant des articles de mode durables et tendance à des prix raisonnables pour encourager davantage de consommateurs à opter pour des choix durables.

- Interpretation2

1. Les deux barres empilées nous montrent comment les consommateurs ne sont pas au courant des impacts de la fast fashion.
Cela peut indiquer un manque de sensibilisation ou de connaissances sur les enjeux environnementaux et sociaux associés à cette industrie; ça souligne l'importance de poursuivre les efforts de sensibilisation sur les impacts négatifs de la fast fashion, tels que l'épuisement des ressources naturelles, la pollution de l'eau et de l'air, les conditions de travail déplorables dans les pays en développement, et la production de déchets textiles.
Il est essentiel d'informer les consommateurs sur les effets de leurs choix de mode sur l'environnement et les populations qui produisent ces vêtements, afin qu'ils puissent faire des choix plus éclairés et conscients. Cela peut aider à encourager les consommateurs à prendre des mesures pour réduire leur impact environnemental et social, telles que l'achat de vêtements durables, l'achat d'occasion, ou le recyclage de leurs vêtements.



2. Enfin, les barres horizontales nous montrent la volonté des consommateurs de passer à une mode plus durable. Cela montre que les consommateurs sont conscients de l'impact de leurs choix de mode sur l'environnement et sont prêts à changer leurs habitudes de consommation.
Cependant, il est important de noter que le passage à une mode plus durable peut être difficile pour certains consommateurs, car cela peut impliquer un changement de comportement, de style de vie et de budget. Il est donc important de soutenir les consommateurs dans leur transition vers une mode plus durable en leur fournissant des informations sur les options durables disponibles, ainsi que des conseils pratiques sur la manière de les intégrer dans leur vie quotidienne.
Il est également important que les entreprises de la mode proposent des options durables à des prix abordables, afin de rendre les choix durables accessibles à tous les consommateurs. Les entreprises peuvent également jouer un rôle important en éduquant les consommateurs sur les choix de mode durables et en les incitant à passer à des alternatives plus durables.

- Interpretation

Cette dernière photo représente des solutions orientées IA et développement durable pour une passer à une mode plus éthique.

Les solutions orientées IA sont les suivantes:
- Utilisation des chat bots pour éduquer les consommateurs sur les impacts de la mode rapide et leur fournir des alternatives éthiques
- Création d'un moteur de recommandation basé sur l'intelligence artificielle pour aider les consommateurs à trouver des produits de mode éthiques adaptés à leur style et à leurs préférences. 
- Utilisation des algorithmes d'apprentissage automatique pour surveiller les prix des produits de mode éthiques et les comparer aux produits de fast fashion afin de montrer aux consommateurs qu'ils peuvent économiser de l'argent et faire une différence positive en achetant des produits éthiques. 
- Développement des systèmes de points de récompense basés sur l'IA pour encourager les consommateurs à acheter des produits de mode éthiques et à encourager leurs amis et leur famille à le faire
- Utilisation des réseaux neuronaux pour créer des images de produits de mode éthiques plus attrayantes et plus engageantes que les produits de fast fashion.
- Développement des applications mobiles qui permettent aux consommateurs d’accéder à des informations sur les produits de mode éthiques et de trouver des produits éthiques dans leur région


Les solutions orientées développement durable sont les suivantes:
- Utiliser des matériaux naturels plutôt que des matériaux synthétiques pour la fabrication des vêtements ? 
- Réutiliser des vêtements anciens pour en créer de nouveaux ? 
- Utiliser des matériaux recyclés pour la fabrication des accessoires de mode ?
- Consommer (beaucoup) moins et d’occasion
- Louer ses vêtements pour moins polluer et  Donner une seconde vie à ses vieux vêtements
- Des vêtements et accessoires de mode à base de bouteilles en plastique
- Offrir des remises pour l’achat de produits éthiques et durables et  Organiser des défilés de mode éthique afin de sensibiliser à la mode responsable.

L'ensemble des solutions proposées ont été bien notés dans l'ensemble et cela peut contribuer à sensibiliser les consommateurs à la mode durable, à encourager l'industrie de la mode à adopter des pratiques plus responsables et à promouvoir un mode de vie plus respectueux de l'environnement.
